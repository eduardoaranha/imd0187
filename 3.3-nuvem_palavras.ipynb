# Tratando as stopwords
---

```python
# Visualizando lista de stop-words em inglês
import nltk
nltk.download('stopwords')
english_stopwords = set(nltk.corpus.stopwords.words('english'))
print(english_stopwords)
```

```python
# Visualizando lista de stop-words em português
pt_stopwords = set(nltk.corpus.stopwords.words('portuguese'))
print(pt_stopwords)
```

```python
# Retirando as stopwords e colocando tudo em minúsculo
tokens = nltk.word_tokenize(text)
content_tokens = [token for token in tokens if token.lower()
 not in pt_stopwords]
print(content_tokens)
```

```python
filtered_sentence = ''
 
for w in tokens:
    if w not in pt_stopwords:
        filtered_sentence = filtered_sentence + ' ' + w
        
wordcloud2 = WordCloud(width=480, height=480, margin=0).generate(filtered_sentence)
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.savefig('nuvem_1.png', dpi = 300)
plt.show()
```

```python
from nltk import FreqDist
FreqDist(content_tokens)
```

# Personalizando stopwords
---

```python
# Personalizando a lista de stopwords
pt_stopwords2 = set(nltk.corpus.stopwords.words('portuguese'))
#pt_stopwords.remove('você')
pt_stopwords2.add(',')
pt_stopwords2.add('.')
print(pt_stopwords2)
```

```python
# Retirando stopwords personalizada
content_tokens = [token for token in tokens if token.lower()
 not in pt_stopwords2]
print(content_tokens)
```

```python
filtered_sentence2 = ''
 
for w in tokens:
    if w not in pt_stopwords2:
        filtered_sentence2 = filtered_sentence2 + ' ' + w
        
wordcloud2 = WordCloud(width=480, height=480, margin=0).generate(filtered_sentence2)
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.savefig('nuvem_2.png', dpi = 300)
plt.show()
```
