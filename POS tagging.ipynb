```python
text = '''Trata-se de agravo interno interposto por ADELMO JOSÉ MARTINS VIEIRA contra decisão de
minha lavra (fls. 433/443), integrada pelo decisum de fls. 457/460, a qual negou
provimento ao seu recurso especial.'''
```

```python
import spacy
from spacy import displacy
nlp = spacy.load("pt_core_news_sm")
```

```python
# Spacy POS tagging:
#    "ADJ": "adjective",
#    "ADP": "adposition",
#    "ADV": "adverb",
#    "AUX": "auxiliary",
#    "CONJ": "conjunction",
#    "CCONJ": "coordinating conjunction",
#    "DET": "determiner",
#    ...
# Veja lista completa em https://github.com/explosion/spaCy/blob/master/spacy/glossary.py
doc = nlp(text)
```

```python
for token in doc:
    print(token.text)
    ind = '    '
    print(ind, token.pos_)
    print(ind, token.tag_)
    print(ind, token.dep_)
    print(ind, spacy.explain(token.pos_))
    print(ind, spacy.explain(token.tag_))
```

```python
spacy.displacy.render(doc)
```

```python
sentence_spans = list(doc.sents)
displacy.render(sentence_spans, style="dep")
```

```python
import nltk
from nltk import pos_tag
from nltk import word_tokenize
```

```python
#nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
```

```python
tokenizer = word_tokenize(text)
pos_tag(tokenizer)
```

```python
sentence = "the little yellow dog barked at the cat"
tokens = nltk.word_tokenize(sentence)
print(tokens)

tag = nltk.pos_tag(tokens)
print(tag)
grammar = "NP: {<DT>?<JJ>*<NN>}"

cp = nltk.RegexpParser(grammar)
result = cp.parse(tag)
print(result)
```
